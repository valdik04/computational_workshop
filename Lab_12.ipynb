```python
import time
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def f(x):
    return x*x - 5*x + 5
```

```python
def df(x):
    return 2*x - 5
```

```python
def f_2(x):
    return x**2

def df_2(x):
    return 2*x
```

```python
def f_3(x):
    return np.log(x+(1+x*x)**(1/2)) + x*x
def df_3(x):
    return x/((x*x+1)**(1/2)*((x*x+1)**(1/2) + x)) + 1/((x*x+1)**(1/2) + x) + 2*x
```

```python
print("Функция: x^2 - 5*x + 5")
eps = 0.00001
xx = -100      # начальное значение
lmb = 0.4  # шаг сходимости
x_plt = np.arange(-100, 100.0, 0.1)
f_plt = [f(x) for x in x_plt]
fig, ax = plt.subplots()    
ax.grid(True)   
ax.plot(x_plt, f_plt)                   
point = ax.scatter(xx, f(xx), c='red')  
xx_1 = xx - lmb*df(xx)
step = 1
mn = 100
while abs(xx_1-xx) > eps:
    #eps = eps/min(step+1, mn)
    ax.scatter(xx_1, f(xx_1), c='red')  
    xx = xx_1
    xx_1 = xx_1 - lmb*df(xx_1)    
    step = step + 1
    


print("Точка минимума:", xx)
print("Количество шагов:", step)
ax.scatter(xx, f(xx), c='blue')
plt.show()
```

```python
print("Функция: x^2")
eps = 0.00000001
xx = 50      # начальное значение
lmb = 0.3  # шаг сходимости
x_plt = np.arange(-50.0, 50.0, 0.1)
f_plt = [f_2(x) for x in x_plt]
fig, ax = plt.subplots()    
ax.grid(True)   
ax.plot(x_plt, f_plt)                   
point = ax.scatter(xx, f_2(xx), c='red')  
xx_1 = xx - lmb*df_2(xx)
step = 1
mn = 100
while abs(f_2(xx_1)-0) > eps:
    ax.scatter(xx_1, f_2(xx_1), c='red')  
    xx = xx_1
    xx_1 = xx_1 - lmb*df_2(xx_1)    
    step = step + 1
    


print("Точка минимума:", xx)
print("Количество шагов:", step)
ax.scatter(xx, f_2(xx), c='blue')
plt.show()
```

```python
print("Функция: x^2 - 5*x + 5")
eps = 0.00000001
k = 0
a = 1
a_= 1
y = 150
x_ = y #начальное приближение
x_plt = np.arange(-150.0, 150.0, 0.1)
f_plt = [f(x) for x in x_plt]
fig, ax = plt.subplots()    
ax.grid(True)   
ax.plot(x_plt, f_plt)                   
point = ax.scatter(y, f(y), c='red')  
z = random.uniform(-150.0, 150.0)
x = x_
lmb = abs((x - z)/(df(x) - df(z)))

y_ = y - 10
k = 0
while abs(y_ - y) > eps:
    y_ = y
    i = 0
    while (f(y) - f(y-2**(-i)*lmb*df(y)) < 2**(-i-1)*lmb*abs(df(y))*abs(df(y))):
        i = i + 1
    lmb = 2**(-i)*lmb
    x = y - lmb*df(y)
    a = (1 + (4*a*a+1)**(1/2))/2
    y = x + (a_-1)*(x-x_)/a
    a_ = a
    x_ = x
    ax.scatter(y, f(y), c='red') 
    k = k + 1
print("Точка минимума:", y)
print("Количество шагов:", k)
ax.scatter(y, f(y), c='blue')
plt.show()
```
